### Checking for country that made the highest sales and profit

##### Sales

```python
maxsales = dt.groupby("Country")["Sales"].max()
```

```python
maxsales.index
```

```python
maxsales.values
```

```python
maxsales = maxsales.to_frame()
```

```python
maxsales = maxsales.sort_values(by="Sales", ascending=False).head(10)
```

```python
maxsales
```

```python
plt.bar(maxsales.index, maxsales["Sales"])
plt.title("Top 5 Sales by Country")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()
```

##### Profit

```python
maxprofit = dt.groupby("Country")["Profit"].max()
```

```python
maxprofit = maxprofit.to_frame()
```

```python
maxprofit = maxprofit.sort_values(by="Profit", ascending=False).head()
```

```python
maxprofit
```

```python
plt.barh(maxprofit.index, maxprofit["Profit"],color=['gold','green','blue','red','pink']
)
plt.title("Top 5 Countries by Profit")
plt.xlabel("Country")
plt.ylabel("Profit")
plt.show()
```

### State with the highest profit and sales

###### Profit

```python
stateprofit = dt.groupby('State')['Profit'].sum()
stateprofit = stateprofit.to_frame()
stateprofit = stateprofit.sort_values(by="Profit", ascending=False).head()
stateprofit
```

```python
plt.pie(stateprofit["Profit"],autopct = '%.1f%%',labels=stateprofit.index ); 
```

##### Sales 

```python
statesales = dt.groupby('State')['Sales'].sum()
statesales = statesales.to_frame()
statesales = statesales.sort_values(by="Sales", ascending=False).head()
statesales
```

```python
plt.pie(statesales["Sales"],autopct = '%.1f%%',labels=statesales.index ); 
```
